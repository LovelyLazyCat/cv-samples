# 基于特征的模型训练

本代码基于提取的特征进行模型训练(特征提取详见`extract_features.ipynb`).

我们可以基于多个特征进行训练,这样的训练效果往往更佳(可以达到Kaggle Dog Vs Cat竞赛前10%).

下面的代码中,采用了以下四种组合特征的方式,你可以任意组合(前提是你提取了这个网络的特征).

```python
from train import TrainConfiguration, train
from utils import plot_metrics
```

## Inceptin 单特征训练

Inception单特征训练用作基准模型,用来对比其他别组合特征训练的效果

```python
conf = TrainConfiguration()

# 本次训练的名称,在保存模型和稍后进行测试的时候需要用到
conf.name = 'inception'

# 特征的路径,注意这里是训练特征的路径
conf.features_path = r'/usr/local/data/cat_dog/features/train'

# 模型的保存路径,请填写一个已经创建的空目录
conf.save_path = r'/usr/local/saved_models/cat_dog'

# 训练轮次
conf.epochs = 120

# 使用哪些特征进行训练,这里只使用Inception
conf.train_features = ['inception']

# 验证集大小
conf.valid_size = 100

conf.show_pre_epoch = 20
```

```python
train_losses, train_acc, valid_losses, valid_acc = train(conf)
```

绘制Loss曲线

```python
plot_metrics(train=train_losses, valid=valid_losses, title='Inception Losses', xlabel='step', ylabel='loss', step=80)
```

绘制Accuracy 曲线

```python
plot_metrics(train=train_acc, valid=valid_acc, title='Inception Accuracy', xlabel='step', ylabel='acc', step=80)
```

绘制Accuracy曲线

```python
plot_metrics(train=train_acc, valid=valid_acc, title='Inception+Xception Accuracy', xlabel='step', ylabel='acc', step=80)
```
