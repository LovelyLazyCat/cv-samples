# 图像的膨胀以及腐蚀

膨胀和腐蚀被称为**形态运算**.膨胀通过向图像中的对象边缘增加像素, 放大亮白区域; 腐蚀刚好相反, 它删除对象边缘的像素, 并缩小对象的大小.

```python
import utils

import cv2

import matplotlib.pyplot as plt

import numpy as np
```

我们以一张简单的字母图像为例子来看看膨胀和腐蚀的效果

```python
image = utils.imread('images/j.png')

gray = utils.rgb2gray(image)

plt.imshow(gray, cmap='gray')
```

要在OpenCV中膨胀图像,可以使用`dilate`函数和三个输入: 原始二元图像,确定膨胀大小的内核(均不会导致默认大小)以及进行膨胀的迭代次数(通常为 1).在下面的示例中,我们用到一个 5x5 内核,它像过滤器一样在图像上移动,如果在 5x5 窗口中,像素周围的任何像素是白色,则使该像素变成白色!

```python
kernel = np.ones((5, 5), np.uint8)

dilation = cv2.dilate(image, kernel, iterations=1)

plt.imshow(dilation, cmap='gray')
```

下面进行腐蚀图像,留意输出和膨胀的区别.

```python
erosion = cv2.erode(image, kernel, iterations=1)
plt.imshow(erosion)
```
